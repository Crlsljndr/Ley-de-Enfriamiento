```python
import math
from pylab import *
global k
global T
def f(x,y): 
    global k
    global T
    return (-k*(y-T)) # Funcion de EDO

def y(x):
    global k
    global T
    return (8 * math.e**(-k*x))+T # Solución exacta EDO con PVI

def rungekutta(f,x,y,h,m):
    u=[]
    v=[]
    for i in range(m):
        k1 = h*f(x,y)
        k2 = h*f(x+h/2,y+k1/2)
        k3 = h*f(x+h/2,y+k2/2)
        k4 = h*f(x+h,y+k3)
        y = y+1/6*(k1+2*k2+2*k3+k4)
        x = x+h
        u = u+[x]
        v = v+[y]
        
    return [u,v]
k = 0.2877
T = 22
[u,v] = rungekutta(f,0,30,1,20)
plot(u,v,'or')
x = arange(0,21,1)
plot(x,y(x),'ob')
grid(True)
show()
```
